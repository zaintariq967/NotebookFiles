### List in Python

```python
nums =[1,2,3,4]

# LIST OPERATIONS

# add single item to end o list
nums.append(10)
print(nums)

# multiple items add in list (adding another list inside list)
nums.extend([10,15])
print(nums)

# tells insert element at specific insert like in below insert element 50 at index 1 
nums.insert(1,50)
print(nums)

# remove that value from list
nums.remove(50)

# remove last element from list
nums.pop()
print(nums)

# remve by index
nums.pop(1)
print(nums)

# removes complete list: empty
nums.clear()
print(nums)
```

```python
# find index
nums.index(3)
print(nums)

# count occurences
nums.count(2)
print(nums)

# reverse list
nums.reverse()
print(nums)

# sort ascending 
nums.sort()
print(nums)

# sort descending
nums.sort(reverse=True)
print(nums)

# copy list
copy_nums = nums.copy()
print(nums)
```

```python
try:
    x = int("abc")          # ValueError
    print(10 / 0)           # ZeroDivisionError
    print(lst[5])           # IndexError
    print(d['b'])           # KeyError
    open("data.txt")        # FileNotFoundError
except ValueError:
    print("Invalid value.")
except ZeroDivisionError:
    print("Cannot divide by zero.")
except IndexError:
    print("List index is out of range.")
except KeyError:
    print("Dictionary key not found.")
except FileNotFoundError:
    print("File does not exist.")
```

### Exception Handling

```python
try:
    age=int(input("Enter AGE: "))
except:
    print("kindly add digits")
```

```python
try:
    a=10
    b=int(input("Enter a number : "))
    c=a/b
    print("Result:",c)
except ZeroDivisionError:
    print("Cannot divide by zero!")
except ValueError:
    print("Input input! Please enter a number")
except Exception as e:
    print("An error occurred:",c)
else:
    print("No error. Program executed successfully.")
finally:
    print("This block always executes")
```
