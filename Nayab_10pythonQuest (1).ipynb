### Student Grade Calculator

```python
def calculate_grade(percentage):
    if percentage >= 90:
        return "A"
    elif percentage >= 80:
        return "B"
    elif percentage >= 70:
        return "C"
    elif percentage >= 60:
        return "D"
    return "F"
```

```python
marks = []

for i in range(1, 6):
    value = float(input(f"Enter marks of subject {i}: "))
    marks.append(value)

total = sum(marks)
average = total / len(marks)
percentage = (total / 500) * 100

grade = calculate_grade(percentage)

print("Total:", total)
print("Average:", average)
print("Percentage:", round(percentage, 2))
print("Grade:", grade)
```

### Number Analyzer

```python
def check_sign(number):
    if number > 0:
        return "Positive"
    elif number < 0:
        return "Negative"
    return "Zero"


def check_even_odd(number):
    if number % 2 == 0:
        return "Even"
    return "Odd"
```

```python
num = int(input("Enter a number: "))

print(check_sign(num))
print(check_even_odd(num))
```

### Simple Calculator Using Functions

```python
def add(a, b):
    return a + b


def subtract(a, b):
    return a - b


def multiply(a, b):
    return a * b


def divide(a, b):
    if b == 0:
        return "Cannot divide by zero"
    return a / b
```

```python
print("1. Addition")
print("2. Subtraction")
print("3. Multiplication")
print("4. Division")

choice = input("Choose operation: ")

first = float(input("Enter first number: "))
second = float(input("Enter second number: "))

if choice == "1":
    print(add(first, second))
elif choice == "2":
    print(subtract(first, second))
elif choice == "3":
    print(multiply(first, second))
elif choice == "4":
    print(divide(first, second))
else:
    print("Invalid choice")
```

### List Statistics 

```python
def show_statistics(values):
    print("Largest:", max(values))
    print("Smallest:", min(values))
    print("Sum:", sum(values))
    print("Average:", sum(values) / len(values))

    unique = sorted(set(values), reverse=True)

    if len(unique) > 1:
        print("Second Largest:", unique[1])
    else:
        print("Second Largest does not exist")
```

```python
numbers = input("Enter numbers separated by space: ").split()

data = []

for item in numbers:
    data.append(int(item))

show_statistics(data)
```

### Student Record Management

```python
def add_student(records):
    name = input("Student name: ")
    marks = float(input("Marks: "))
    records[name] = marks


def update_student(records):
    name = input("Student name: ")

    if name in records:
        records[name] = float(input("New marks: "))
    else:
        print("Student not found")


def delete_student(records):
    name = input("Student name: ")

    if name in records:
        del records[name]
    else:
        print("Student not found")


def search_student(records):
    name = input("Student name: ")

    if name in records:
        print(name, ":", records[name])
    else:
        print("Student not found")


def display_records(records):
    if len(records) == 0:
        print("No records available")
    else:
        for name, marks in records.items():
            print(name, "-", marks)
```

```python
students = {}

while True:

    print("\n1. Add Student")
    print("2. Update Marks")
    print("3. Delete Student")
    print("4. Search Student")
    print("5. Display Records")
    print("6. Exit")

    option = input("Enter choice: ")

    if option == "1":
        add_student(students)

    elif option == "2":
        update_student(students)

    elif option == "3":
        delete_student(students)

    elif option == "4":
        search_student(students)

    elif option == "5":
        display_records(students)

    elif option == "6":
        break

    else:
        print("Invalid choice")
```

```python
def count_words(sentence):

    words = sentence.lower().split()
    result = {}

    for word in words:
        if word in result:
            result[word] += 1
        else:
            result[word] = 1

    return result
```

```python
text = input("Enter a sentence: ")

frequency = count_words(text)

print("\nWord Frequency")

for word, count in frequency.items():
    print(word, ":", count)
```

```python
def check_password(password):

    length = len(password) >= 8

    upper = False
    lower = False
    digit = False
    special = False

    symbols = "!@#$%^&*()-_=+[]{}|\\:;\"'<>,.?/"

    for ch in password:

        if ch.isupper():
            upper = True

        elif ch.islower():
            lower = True

        elif ch.isdigit():
            digit = True

        elif ch in symbols:
            special = True

    return length and upper and lower and digit and special
```

```python
pwd = input("Enter password: ")

if check_password(pwd):
    print("Valid Password")
else:
    print("Invalid Password")
```

```python
def show_cart(cart, prices):

    total = 0

    if len(cart) == 0:
        print("Cart is empty")
        return

    print("\nItems in Cart")

    for item, quantity in cart.items():
        bill = prices[item] * quantity
        total += bill
        print(item, "-", quantity, "-", bill)

    print("Total Bill:", total)
```

```python
prices = {
    "Apple": 120,
    "Banana": 60,
    "Milk": 180,
    "Bread": 90,
    "Eggs": 25
}

cart = {}

while True:

    print("\n1. Add Item")
    print("2. Remove Item")
    print("3. Update Quantity")
    print("4. Display Cart")
    print("5. Exit")

    choice = input("Enter choice: ")

    if choice == "1":

        item = input("Item name: ")

        if item in prices:
            quantity = int(input("Quantity: "))
            cart[item] = quantity
        else:
            print("Item not available")

    elif choice == "2":

        item = input("Item name: ")

        if item in cart:
            del cart[item]
        else:
            print("Item not found")

    elif choice == "3":

        item = input("Item name: ")

        if item in cart:
            quantity = int(input("New quantity: "))
            cart[item] = quantity
        else:
            print("Item not found")

    elif choice == "4":

        show_cart(cart, prices)

    elif choice == "5":

        break

    else:

        print("Invalid Choice")
```

```python
def divide_numbers():

    try:

        first = float(input("Enter first number: "))
        second = float(input("Enter second number: "))

        answer = first / second

    except ValueError:

        print("Please enter valid numbers.")

    except ZeroDivisionError:

        print("Division by zero is not allowed.")

    else:

        print("Answer:", answer)

    finally:

        print("Program Finished")
```

```python
def add_book():

    name = input("Book name: ")

    if name not in books:
        books[name] = "Available"
        print("Book Added")
    else:
        print("Book already exists")


def search_book():

    name = input("Book name: ")

    if name in books:
        print(name, "-", books[name])
    else:
        print("Book not found")


def borrow_book():

    name = input("Book name: ")

    if name in books:

        if books[name] == "Available":
            books[name] = "Borrowed"
            print("Book Borrowed")
        else:
            print("Book already borrowed")

    else:
        print("Book not found")


def return_book():

    name = input("Book name: ")

    if name in books:

        if books[name] == "Borrowed":
            books[name] = "Available"
            print("Book Returned")
        else:
            print("Book is already available")

    else:
        print("Book not found")


def display_books():

    print("\nLibrary Books")

    available = False

    for book, status in books.items():

        if status == "Available":
            print(book)
            available = True

    if not available:
        print("No books available")
```

```python
books = {}

while True:

    print("\n1. Add Book")
    print("2. Search Book")
    print("3. Borrow Book")
    print("4. Return Book")
    print("5. Display Available Books")
    print("6. Exit")

    option = input("Choose option: ")

    if option == "1":

        add_book()

    elif option == "2":

        search_book()

    elif option == "3":

        borrow_book()

    elif option == "4":

        return_book()

    elif option == "5":

        display_books()

    elif option == "6":

        break

    else:

        print("Invalid Option")
```
