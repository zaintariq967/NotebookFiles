# 📁 PART 1: File Handling
Python can **create, read, write, and delete** files easily.

## 1️⃣ Open Modes
| Mode | Meaning |
|---|---|
| `'r'` | Read (default) |
| `'w'` | Write (creates new / overwrites) |
| `'a'` | Append (add to end) |
| `'x'` | Create new file |
| `'rb'` | Read binary |

## 2️⃣ Write to a File

```python
f = open("test.txt", "w")
f.write("Hello, Aamir!")
f.close()
print("File written!")
```

## 3️⃣ Read a File

```python
f = open("test.txt", "r")
content = f.read()
f.close()
print(content)
```

## 4️⃣ Write Multiple Lines

```python
f = open("lines.txt", "w")
f.write("Line 1\n")
f.write("Line 2\n")
f.write("Line 3\n")
f.close()
print("Done!")
```

## 5️⃣ Read Line by Line

```python
f = open("lines.txt", "r")
for line in f:
    print(line.strip())
f.close()
```

## 6️⃣ `readline()` — Read One Line

```python
f = open("lines.txt", "r")
print(f.readline())  # first line
print(f.readline())  # second line
f.close()
```

## 7️⃣ `readlines()` — All Lines as a List

```python
f = open("lines.txt", "r")
all_lines = f.readlines()
f.close()
print(all_lines)
```

## 8️⃣ Append to a File
`'a'` mode adds to the end — does NOT delete old content.

```python
f = open("test.txt", "a")
f.write("\nNew line added!")
f.close()

f = open("test.txt", "r")
print(f.read())
f.close()
```

## 9️⃣ Best Way: `with` Statement
Automatically closes the file. No need to call `.close()`

```python
with open("test.txt", "r") as f:
    content = f.read()
    print(content)
```

## 🔟 Write with `with`

```python
with open("students.txt", "w") as f:
    f.write("Ali\nSara\nUsman\n")

with open("students.txt", "r") as f:
    print(f.read())
```

## 1️⃣1️⃣ Write a List to File using `writelines()`

```python
names = ["Bilal\n", "Ayesha\n", "Zara\n"]

with open("names.txt", "w") as f:
    f.writelines(names)

with open("names.txt", "r") as f:
    print(f.read())
```

## 1️⃣2️⃣ Check if File Exists

```python
import os

if os.path.exists("test.txt"):
    print("File exists ✅")
else:
    print("File NOT found ❌")
```

## 1️⃣3️⃣ Delete a File

```python
import os

if os.path.exists("names.txt"):
    os.remove("names.txt")
    print("File deleted 🗑️")
else:
    print("Nothing to delete")
```

## 1️⃣4️⃣ Rename a File

```python
import os

os.rename("students.txt", "class_list.txt")
print("Renamed ✅")
```

## 1️⃣5️⃣ Get File Size

```python
import os

size = os.path.getsize("test.txt")
print(f"File size: {size} bytes")
```

# Real Time Example

```python
# Step 1: Take user input
name = input("Enter Name: ")
age = input("Enter Age: ")
gender = input("Enter Gender: ")
email = input("Enter Email: ")

# Step 2: Write data into bio.txt in vertical table format
with open("bio.txt", "w") as file:
    file.write("="*40 + "\n")
    file.write(f"{'Name':<10} : {name}\n")
    file.write(f"{'Age':<10} : {age}\n")
    file.write(f"{'Gender':<10} : {gender}\n")
    file.write(f"{'Email':<10} : {email}\n")
    file.write("="*40 + "\n")

print("\nBasic information saved successfully!")

# Step 3: Take phone number input
phone = input("\nEnter Phone Number to append: ")

# Step 4: Append phone number
with open("bio.txt", "a") as file:
    file.write(f"{'Phone':<10} : {phone}\n")
    file.write("="*40 + "\n")

print("\nPhone number appended successfully!")
```

# ⚠️ PART 2: Exception Handling
Prevents program from **crashing** when an error occurs.

## 1️⃣6️⃣ What is an Exception?
An error that happens **while the program is running**.

```python
# This will crash without try-except
# print(10 / 0)   # ZeroDivisionError
# print(int("abc"))  # ValueError
print("Run safely with try-except!")
```

## 1️⃣7️⃣ Basic `try` and `except`

```python
try:
    result = 10 / 0
except:
    print("Error caught! Cannot divide by zero ❌")
```

## 1️⃣8️⃣ Catch Specific Error Type

```python
try:
    x = int("hello")
except ValueError:
    print("That's not a number! 🔢")
```

## 1️⃣9️⃣ `ZeroDivisionError`

```python
try:
    ans = 100 / 0
except ZeroDivisionError:
    print("Cannot divide by zero! ⚠️")
```

## 2️⃣0️⃣ `else` Block
Runs only when **no error** occurred.

```python
try:
    x = 10 / 2
except ZeroDivisionError:
    print("Error!")
else:
    print(f"No error! Answer = {x} ✅")
```

## 2️⃣1️⃣ `finally` Block
**Always runs** — error or not.

```python
try:
    x = 10 / 0
except ZeroDivisionError:
    print("Error caught!")
finally:
    print("This always runs 🔄")
```

## 2️⃣2️⃣ Full Structure: `try` + `except` + `else` + `finally`

```python
try:
    num = int(input("Enter a number: "))
    result = 100 / num
except ValueError:
    print("Not a valid number!")
except ZeroDivisionError:
    print("Can't divide by zero!")
else:
    print(f"Result: {result}")
finally:
    print("Program finished.")
```

## 2️⃣3️⃣ Catch Multiple Exceptions Together

```python
try:
    x = int("abc")
except (ValueError, TypeError):
    print("ValueError or TypeError caught! ⚠️")
```

## 2️⃣4️⃣ Print the Error Message

```python
try:
    x = 5 / 0
except ZeroDivisionError as e:
    print(e)
```

## 2️⃣5️⃣ `FileNotFoundError` — File Doesn't Exist

```python
try:
    f = open("missing.txt", "r")
except FileNotFoundError:
    print("File not found! 📂❌")
```

## 2️⃣6️⃣ `IndexError` — List Index Out of Range

```python
nums = [1, 2, 3]

try:
    print(nums[10])
except IndexError:
    print("Index out of range! 📋")
```

## 2️⃣7️⃣ `KeyError` — Dictionary Key Missing

```python
data = {"name": "Ali", "age": 20}

try:
    print(data["city"])
except KeyError:
    print("Key does not exist! 🔑")
```

## 2️⃣8️⃣ `TypeError` — Wrong Data Type

```python
try:
    result = "5" + 5
except TypeError:
    print("Cannot add string and number! 🚫")
```

## 2️⃣9️⃣ `raise` — Manually Throw an Error

```python
age = -5

try:
    if age < 0:
        raise ValueError("Age cannot be negative!")
except ValueError as e:
    print(f"Error: {e}")
```

## 3️⃣0️⃣ Custom Exception Class

```python
class AgeError(Exception):
    pass

try:
    age = 200
    if age > 150:
        raise AgeError("Age is too high!")
except AgeError as e:
    print(f"Custom Error: {e} 🚨")
```

## 3️⃣1️⃣ File Handling + Exception Handling Together

```python
filename = "test.txt"

try:
    with open(filename, "r") as f:
        print(f.read())
except FileNotFoundError:
    print(f"'{filename}' not found!")
finally:
    print("File operation complete.")
```

## 3️⃣2️⃣ Safe Input — Ask Until Correct

```python
while True:
    try:
        num = int(input("Enter a number: "))
        print(f"You entered: {num}")
        break
    except ValueError:
        print("Invalid! Please enter a number.")
```

## 3️⃣3️⃣ Common Exception Types Summary

| Exception | When it happens |
|---|---|
| `ValueError` | Wrong value type (`int("abc")`) |
| `ZeroDivisionError` | Divide by zero |
| `FileNotFoundError` | File missing |
| `IndexError` | List index too big |
| `KeyError` | Dict key missing |
| `TypeError` | Wrong data type |
| `AttributeError` | Object has no attribute |
| `NameError` | Variable not defined |

## 3️⃣4️⃣ `NameError` — Variable Not Defined

```python
try:
    print(score)
except NameError:
    print("Variable 'score' is not defined! 🔍")
```

## 3️⃣5️⃣ Real Project: Student Grade File System 📊

```python
students = [
    "Ali,85",
    "Sara,92",
    "Usman,78"
]

# Write grades to file
with open("grades.txt", "w") as f:
    for s in students:
        f.write(s + "\n")

# Read and display grades
try:
    with open("grades.txt", "r") as f:
        print("📊 Student Grades:")
        for line in f:
            name, grade = line.strip().split(",")
            print(f"  {name}: {grade}")
except FileNotFoundError:
    print("grades.txt not found!")
except Exception as e:
    print(f"Error: {e}")
```

## ✅ Quick Summary

**File Handling:**
- `open(file, mode)` → open a file
- `read()`, `readline()`, `readlines()` → read
- `write()`, `writelines()` → write
- `with open(...) as f:` → best practice
- `os.remove()`, `os.rename()`, `os.path.exists()` → manage files

**Exception Handling:**
- `try` → code that might fail
- `except` → handle the error
- `else` → runs if no error
- `finally` → always runs
- `raise` → throw your own error
