## 7. Operators in Python

**Operators** are special symbols to perform operations on variables and values.

| Category | Operators |
|----------|-----------|
| Arithmetic | `+`, `-`, `*`, `/`, `//`, `%`, `**` |
| Comparison | `==`, `!=`, `>`, `<`, `>=`, `<=` |
| Assignment | `=`, `+=`, `-=`, `*=`, `/=` |
| Logical | `and`, `or`, `not` |
| Membership | `in`, `not in` |
| Identity | `is`, `is not` |

```python
# Arithmetic Operators
a = 15
b = 4

print("Addition       :", a + b)   # 19
print("Subtraction    :", a - b)   # 11
print("Multiplication :", a * b)   # 60
print("Division       :", a / b)   # 3.75
print("Floor Division :", a // b)  # 3  (removes decimal)
print("Modulus        :", a % b)   # 3  (remainder)
print("Exponent       :", a ** b)  # 50625 (15 to the power 4)
```

```python
# Comparison Operators (always return True or False)
x = 10
y = 20

print("Equal            :", x == y)   # False
print("Not Equal        :", x != y)   # True
print("Greater Than     :", x > y)    # False
print("Less Than        :", x < y)    # True
print("Greater or Equal :", x >= y)   # False
print("Less or Equal    :", x <= y)   # True
```

```python
# Assignment Operators
n = 10
print("Initial value  :", n)

n += 5   # same as n = n + 5
print("After += 5     :", n)

n -= 3   # same as n = n - 3
print("After -= 3     :", n)

n *= 2   # same as n = n * 2
print("After *= 2     :", n)

n /= 4   # same as n = n / 4
print("After /= 4     :", n)
```

```python
# Logical Operators
p = True
q = False

print("p and q :", p and q)  # False (both must be True)
print("p or  q :", p or q)   # True  (at least one must be True)
print("not  p  :", not p)    # False (reverses the value)
print("not  q  :", not q)    # True
```

```python
# Membership Operators
fruits = ['apple', 'banana', 'cherry']

print("'apple' in fruits    :", 'apple' in fruits)      # True
print("'grape' not in fruits:", 'grape' not in fruits)  # True

# Identity Operators
a = [1, 2, 3]
b = [1, 2, 3]
c = a

print("a is c  :", a is c)   # True  (same object in memory)
print("a is b  :", a is b)   # False (same content, different object)
print("a == b  :", a == b)   # True  (same content)
```
