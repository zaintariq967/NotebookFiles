## Function
A function in programming is a self-contained, reusable block of code designed to perform a specific, isolated task. 

- **Reusability:** Written once, a function can be called multiple times across the program.
- **Scope Isolation:** Variables declared inside a function are local to its own workspace.
- **Abstraction:** The internal logic remains a "black box" hidden from the rest of the application.
- **Control Flow Hijacking:** Calling a function pauses the main code, jumps to the function, and returns after completion.

`def function_name(param)` is used to define function.

### User Defined Function

```python
# function defination

def greet():
    return ("Welcome")
```

```python
# Function call

greet()
greet()
greet()
```

```python
# function with arguments 

def greet(name,age):
    print(f"Assalam u Alaikum {name}, Your Age is {age}")
```

```python
greet("Zain",23)
```

```python
greet(23,"zain")
```

```python
# function with arguments 
# postional arguments : if we change pos of argument then will get alternate results e.g:

def greet(name):
    print(f"Assalam u Alaikum {name}")
```

```python
greet("Zain")
```

```python
# keyword arguments: we define value when calling fn , if parameter position is changed we cant get positional error

def greet(name,age=23):
    print("AOA ",name, "I am ",age, " years old ")
    
greet(name="Zain",age=23)
```

```python
# default argument

def greet(name,age=23):
    print("AOA",name,"I am",age, "years old")
```

```python
greet("Zain")
```

```python
# variable length argument fns: vla

def vla(*args):
    print(args)
```

```python
vla(1,2,3)
```

```python
def add(*args):
    print(sum(args))
```

```python
add(1,10,25,56,4,5)
```

```python
# variable length keyword arguments

def data(**args):
    print(args)
```

```python
data(name="Ali")
```

```python
data(name="Ali", age=23)
```

```python
data(name="Ali", age=23, marks=85)
```
