List Indexing and Slicing

```python
# indexing start at 0 give index , same applies to slicing 
nums=[1,3,5,6,55,4,7,8,9,5,56,10,56,9,10,6,7]
nums[6] # index 6 start from 0 
nums[:] # all print 
nums[2:7] # start from 2nd index till 7 index
nums[7:] # from 7 index to onward 
nums[:7]
```

```python
lst=[2,5,10]
lst.append(7)
print(lst)

lst.insert(1,10)
print(lst)

lst.remove(2)
print(lst)

lst.pop(1)
print(lst)

lst.clear()



```
