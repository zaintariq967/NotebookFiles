## TASK10: BMI CATEGORY

```python
# TASK 10

weight = 70      # in kilograms
height = 1.75    # in meters

bmi = weight / (height * height)

print("BMI =", bmi)

if bmi < 18.5:
    print("Underweight")
elif bmi >= 18.5 and bmi <= 24.9:
    print("Normal")
elif bmi >= 25 and bmi <= 29.9:
    print("Overweight")
else:
    print("Obese")
```

## TASK 07: PRIME Number

```python
# TASK 07

x = int(input("Enter a Number: "))
prime = True

for i in range(2, x):
    if x % i == 0:
        prime = False
        break

if prime:
    print("Number is Prime")
else:
    print("Not Prime (Composite)")
```

## TASK 08: FIBONACCI SERIES

```python
# TASK 08

n = int(input("Enter Number of Terms: "))

a = 0
b = 1

for i in range(n):
    print(a, end=" ")
    c = a + b
    a = b
    b = c
```
