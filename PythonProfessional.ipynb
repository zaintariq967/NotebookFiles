### Modification to it : Adding Heading Introduction to AI at top of above para in same file 

### Step 1: Read the file

```python
with open("files/ai.txt", "r") as f:
    content = f.read()
```

### Step 2: Write the heading + old content

```python
with open("files/ai.txt", "r+") as f:

    # Read existing content
    content = f.read()

    # Cursor position after reading
    print("Position after read():", f.tell())

    # Move cursor to the beginning
    f.seek(0)
    print("Position after seek(0):", f.tell())

    # Write heading + old content
    f.write("Introduction to AI\n\n" + content)

    # Move cursor back to the beginning
    f.seek(0)

    # Display updated file
    print("\nUpdated File Content:\n")
    print(f.read())
```

### WORK WITH CSV Files

<div style="background:#E8F5E9;padding:15px;border-left:6px solid #2E7D32;border-radius:8px;">

## Step 1: Create Sample Data

In this step, we create a **2D list** containing sample student records.

- The first row contains the **column headers**.
- Each remaining row represents one student's information.
- This data will later be written into a CSV file.

</div>

```python
import csv
data=[
    ['Name','Age','City'],
    ['Aamir','25','Islamabad'],
    ['Sara','22','Lahore'],
]
```

<div style="background:#E3F2FD;padding:15px;border-left:6px solid #1565C0;border-radius:8px;">

## Step 2: Write Data to a CSV File

This program creates a new CSV file named **output.csv** inside the **files** folder.

### Working
- `open(..., 'w')` creates or overwrites the CSV file.
- `newline=''` prevents blank lines between rows.
- `encoding='utf-8'` allows storing Unicode characters.
- `csv.writer()` creates a CSV writer object.
- `writer.writerows(data)` writes all rows from the list into the CSV file.

</div>

```python
with open('files\output.csv','w',
          newline='',encoding='utf-8') as file:
    writer=csv.writer(file)
    writer.writerows(data)
```

<div style="background:#FFF8E1;padding:15px;border-left:6px solid #F9A825;border-radius:8px;">

## Step 3: Read Data from the CSV File

This program reads the contents of **output.csv** using the `csv.reader()` function.

### Working
- `open(..., 'r')` opens the CSV file in read mode.
- `csv.reader(file)` reads the file row by row.
- The `for` loop iterates through each row.
- `print(row)` displays every record stored in the CSV file.

</div>

```python
import csv

# Reading CSV File
with open('files\output.csv', 'r', newline='', encoding='utf-8') as file:
    reader = csv.reader(file)

    for row in reader:
        print(row)
```

<div style="background:#F1F8E9;padding:15px;border-left:6px solid #558B2F;border-radius:8px;">

## Output

The CSV file has been created successfully and its contents are displayed above.

Each row is returned as a **list**, where every element represents a separate column value.

</div>

### JSON FILE

```python
import json
data={
    "users":[
        {"name":"Aamir","age":25},
        {"name":"Zain","age":23}
    ]
}

with open('files/output.json','w',
          encoding='utf-8') as file:
    json.dump(data,file,indent=4)
```

```python
import json

with open('files/output.json','r',
          encoding='utf-8') as file:
    data=json.load(file)
    print(data)
    print(data['users'][0],['name'])
```
