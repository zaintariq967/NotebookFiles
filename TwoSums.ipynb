#  Problem Statement

Given a list of integers `nums` and an integer `target`, write a Python function to find the **indices** of the two numbers whose **sum is equal to the target**.

- You may assume that **exactly one solution exists**.
- You **cannot use the same element twice**.
- Return the **indices**, not the numbers.

##  Input

```python
nums = [8, 2, 4, 7, 5]
target = 9
```
## Output

```python
[1, 3]
```


## Solution # 01 (Brute Force)
## Algorithm
Pick the first number.
Add it with every number after it.
If the sum equals the target, return their indices.
Otherwise, move to the next number.
Repeat until you find the answer.

````python
def twoSum(nums, target):

    for i in range(len(nums)):      # Pick first number

        for j in range(i + 1, len(nums)):   # Compare with remaining numbers

            if nums[i] + nums[j] == target:

                return [i, j]
````

````python
%%time
nums=[8,2,4,7,5]
target=9
print(twoSum(nums, target))
````

## Solution # 02 (Optimal Solution)

**Input:** `nums` (list of integers), `target` (integer)

1. Create an **empty dictionary** called `seen`.
2. Traverse the list `nums` from left to right.
3. For each element:

   * Find the **complement** using:

     ```
     complement = target - current_number
     ```
   * Check if the complement is already present in `seen`.

     * If **Yes**, return the indices:

       ```
       [seen[complement], current_index]
       ```
     * If **No**, store the current number and its index in the dictionary:

       ```
       seen[current_number] = current_index
       ```
4. Continue until the pair is found.
5. Return the indices of the two numbers whose sum equals the target.

````python
def twoSum(nums, target):

    seen = {}

    for i in range(len(nums)):

        complement = target - nums[i]

        if complement in seen:
            return [seen[complement], i]

        seen[nums[i]] = i
````

````python
%%time
nums = [8, 2, 4, 7, 5]
target = 9

answer = twoSum(nums, target)

print(answer)
````
