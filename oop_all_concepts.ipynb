## 2️⃣ `__init__` Method (Constructor)
Called automatically when object is created. Assigns values.

```python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age
```

```python
d = Dog("Max", 3)
print(d.name)  # Max
print(d.age)   # 3
```

```python
class std:
    def __init__(self, name, age, gender):
        self.name = name
        self.age = age
        self.gender = gender

    def show_info(self):
        print("My name is ", self.name, ' and i am ', self.age, ' year old')
```

```python
std1 = std('ali', 23, 'M')
std2 = std('ahmed', 24, 'M')
```

```python
std2.show_info()
```

## 3️⃣ Instance Attributes vs Class Attributes
- **Instance**: belongs to one object
- **Class**: shared by all objects

```python
class Car:
    wheels = 4          # class attribute (shared)

    def __init__(self, brand):
        self.brand = brand  # instance attribute

c1 = Car("Toyota")
c2 = Car("Honda")
print(c1.brand, c1.wheels)   # Toyota 4
print(c2.brand, c2.wheels)   # Honda4
```

## 4️⃣ Methods (Functions inside a Class)
`self` is always the first parameter — it refers to the object itself.

```python
class Dog:
    def __init__(self, name):
        self.name = name

    def bark(self):
        print(f"{self.name} says: Woof!")


d = Dog("Buddy")
d.bark()
```

## 5️⃣ `__str__` Method
Controls what prints when you `print(object)`

```python
class Student:
    def __init__(self, name, grade):
        self.name = name
        self.grade = grade

    def __str__(self):
        return f"Student: {self.name}, Grade: {self.grade}"

s = Student("Ali", "A")
```

```python
class std:
    def __init__(self, name, age, gender, email, phy, math, comp):
        self.name = name
        self.age = age
        self.gender = gender
        self.email = email
        self.phy = phy
        self.math = math
        self.comp = comp

    # Calculate total marks
    def calculate_total(self):
        return self.phy + self.math + self.comp

    # Calculate percentage
    def calculate_percentage(self):
        return self.calculate_total() / 3

    # Calculate grade
    def calculate_grade(self):
        percentage = self.calculate_percentage()

        if percentage >= 90:
            return "A+"
        elif percentage >= 80:
            return "A"
        elif percentage >= 70:
            return "B"
        elif percentage >= 60:
            return "C"
        elif percentage >= 50:
            return "D"
        else:
            return "F"

    # Show basic info
    def show_info(self):
        print("Name:", self.name)
        print("Age:", self.age)
        print("Gender:", self.gender)
        print("Email:", self.email)

    # Show result
    def show_result(self):
        print("Physics:", self.phy)
        print("Math:", self.math)
        print("Computer:", self.comp)
        print("Total:", self.calculate_total())
        print("Percentage:", self.calculate_percentage())
        print("Grade:", self.calculate_grade())

    # String representation (Table Format)
    def __str__(self):
        return f"""
+----------------+----------------------+
| Name           | {self.name:<20} |
| Age            | {self.age:<20} |
| Gender         | {self.gender:<20} |
| Email          | {self.email:<20} |
| Physics        | {self.phy:<20} |
| Math           | {self.math:<20} |
| Computer       | {self.comp:<20} |
| Total          | {self.calculate_total():<20} |
| Percentage     | {self.calculate_percentage():<20.2f} |
| Grade          | {self.calculate_grade():<20} |
+----------------+----------------------+
"""


```

```python
std1 = std("ali", 24,'M', 'ali@gmail.com', 87,56,89)
```

```python
print(std1)
```

```python
import math

class Calculas:

    # -------------------------
    # Power Operations
    # -------------------------

    def square(self, x):
        return x ** 2

    def cube(self, x):
        return x ** 3

    def n_power(self, x, n):
        return x ** n

    # -------------------------
    # Roots
    # -------------------------

    def sqrt(self, x):
        return math.sqrt(x)

    def cubert(self, x):
        return x ** (1/3)

    # -------------------------
    # Logarithmic Functions
    # -------------------------

    def logN(self, x):
        return math.log(x)  # Natural log (ln)

    def log10(self, x):
        return math.log10(x)

    # -------------------------
    # Exponential Functions
    # -------------------------

    def en(self, n):
        return math.exp(n)

    def one_over_one_plus_en(self, n):
        return 1 / (1 + math.exp(n))

    def sigmoid(self, x):
        return 1 / (1 + math.exp(-x))

    # -------------------------
    # Hyperbolic Functions
    # -------------------------

    def sinh(self, x):
        return math.sinh(x)

    def cosh(self, x):
        return math.cosh(x)

    def tanh(self, x):
        return math.tanh(x)

    # -------------------------
    # Factorial, Combination, Permutation
    # -------------------------

    def factorial(self, n):
        return math.factorial(n)

    def nCr(self, n, r):
        return math.comb(n, r)

    def nPr(self, n, r):
        return math.perm(n, r)

    # -------------------------
    # Calculus Operations
    # -------------------------

    # Numerical Derivative (Forward Difference)
    def derivative(self, f, x, h=1e-5):
        return (f(x + h) - f(x)) / h

    # Numerical Integration (Trapezoidal Rule)
    def integrate(self, f, a, b, n=1000):
        h = (b - a) / n
        total = 0.5 * (f(a) + f(b))
        for i in range(1, n):
            total += f(a + i * h)
        return total * h

    # Limit Approximation
    def limit(self, f, x, h=1e-5):
        return (f(x + h) + f(x - h)) / 2

    # -------------------------
    # String Representation
    # -------------------------

    def __str__(self):
        return "Calculas - Advanced Calculus Utility Class"
```

```python
calc = Calculas()

# Define function
f = lambda x: x**2

print("Derivative at x=3:", calc.derivative(f, 3))
print("Integral from 0 to 5:", calc.integrate(f, 0, 5))
print("Limit at x=2:", calc.limit(f, 2))
print("Factorial 5:", calc.factorial(5))
print("Sigmoid(2):", calc.sigmoid(2))
```

## 6️⃣ Encapsulation — Hide Data 🔒
Use `__` (double underscore) to make attribute **private**.

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance  # private

    def get_balance(self):
        return self.__balance

    def deposit(self, amount):
        self.__balance += amount

acc = BankAccount(1000)
acc.deposit(500)
print(acc.get_balance())  # 1500
# print(acc.__balance)    # ❌ Error! Private
```

## 7️⃣ Getters and Setters
Control access to private data safely.

```python
class Person:
    def __init__(self, age):
        self.__age = age

    def get_age(self):
        return self.__age

    def set_age(self, age):
        if age > 0:
            self.__age = age
        else:
            print("Invalid age!")

p = Person(20)
p.set_age(25)
print(p.get_age())   # 25
p.set_age(-5)        # Invalid age!
```

## 8️⃣ Inheritance — Child Reuses Parent 👨‍👦
Child class gets all features of the parent class.

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        print(f"{self.name} makes a sound")
```

```python
class Cat(Animal):   # Cat inherits Animal
    def speak(self):
        print(f"{self.name} says: Meow 🐱")

class dog(Animal):
    pass
```

```python
d = dog("husk")
d.speak()
```

## 9️⃣ `super()` — Call Parent Method
Use `super()` to run the parent's `__init__` or methods.

```python
class Vehicle:
    def __init__(self, brand):
        self.brand = brand

class ElectricCar(Vehicle):
    def __init__(self, brand, battery):
        super().__init__(brand)       # call parent __init__
        self.battery = battery

    def info(self):
        print(f"{self.brand} | Battery: {self.battery} kWh ⚡")

e = ElectricCar("Tesla", 100)
e.info()
```

## 🔟 Multiple Inheritance
A class can inherit from **more than one** parent.

```python
class Flyable:
    def fly(self):
        print("I can fly ✈️")

class Swimmable:
    def swim(self):
        print("I can swim 🏊")

class Duck(Flyable, Swimmable):
    pass

d = Duck()
d.fly()
d.swim()
```

## 1️⃣1️⃣ Polymorphism — Same Name, Different Behavior
Same method name works differently in different classes.

```python
class Dog:
    def sound(self):
        print("Woof 🐶")

class Cat:
    def sound(self):
        print("Meow 🐱")

class Cow:
    def sound(self):
        print("Moo 🐮")

animals = [Dog(), Cat(), Cow()]
for a in animals:
    a.sound()   # each calls its own version
```

## 1️⃣2️⃣ Abstraction — Hide Complexity 🎭
Abstract class defines **rules**. Child must implement methods.

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, r):
        self.r = r
    def area(self):
        return 3.14 * self.r ** 2

class Rect(Shape):
    def __init__(self, w, h):
        self.w = w
        self.h = h
    def area(self):
        return self.w * self.h

print(Circle(5).area())   # 78.5
print(Rect(4, 6).area())  # 24
```

## 1️⃣3️⃣ Class Method & Static Method
- `@classmethod` → works with class, not object
- `@staticmethod` → no `self` or `cls` needed

```python
class MathHelper:
    pi = 3.14159

    @classmethod
    def circle_area(cls, r):
        return cls.pi * r * r

    @staticmethod
    def add(a, b):
        return a + b

print(MathHelper.circle_area(5))  # 78.53
print(MathHelper.add(10, 20))     # 30
```

## 1️⃣4️⃣ `@property` Decorator
Access a method like an attribute (no parentheses needed).

```python
class Temperature:
    def __init__(self, celsius):
        self.celsius = celsius

    @property
    def fahrenheit(self):
        return self.celsius * 9/5 + 32

t = Temperature(100)
print(t.fahrenheit)   # 212.0 — no () needed!
```

## 1️⃣5️⃣ Magic / Dunder Methods
Special methods with `__` that Python calls automatically.

```python
class Book:
    def __init__(self, title, pages):
        self.title = title
        self.pages = pages

    def __str__(self):       # print()
        return f"📖 {self.title}"

    def __len__(self):       # len()
        return self.pages

    def __add__(self, other): # + operator
        return self.pages + other.pages

b1 = Book("Python", 300)
b2 = Book("OOP", 200)
print(b1)          # 📖 Python
print(len(b1))     # 300
print(b1 + b2)     # 500
```

## 1️⃣6️⃣ Object Comparison with `__eq__`
Compare objects using `==`.

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __eq__(self, other):
        return self.x == other.x and self.y == other.y

p1 = Point(1, 2)
p2 = Point(1, 2)
p3 = Point(3, 4)
print(p1 == p2)  # True
print(p1 == p3)  # False
```

## 1️⃣7️⃣ Inheritance Chain (Multi-level)
Grandparent → Parent → Child

```python
class LivingThing:
    def breathe(self):
        print("Breathing...")

class Animal(LivingThing):
    def move(self):
        print("Moving...")

class Dog(Animal):
    def bark(self):
        print("Barking! 🐶")

d = Dog()
d.breathe()  # from LivingThing
d.move()     # from Animal
d.bark()     # from Dog
```

## 1️⃣8️⃣ `isinstance()` and `issubclass()`
Check object type or class relationship.

```python
class Animal: pass
class Dog(Animal): pass

d = Dog()
print(isinstance(d, Dog))      # True
print(isinstance(d, Animal))   # True (inherited)
print(issubclass(Dog, Animal)) # True
```

## 1️⃣9️⃣ Method Overriding
Child changes the parent method behavior.

```python
class Employee:
    def salary(self):
        return 50000

class Manager(Employee):
    def salary(self):          # overrides parent
        return 100000

e = Employee()
m = Manager()
print("Employee salary:", e.salary())  # 50000
print("Manager salary:", m.salary())   # 100000

```

## 2️⃣0️⃣ Real-World Example: Library System 📚
Putting it all together!

```python
class Library:
    def __init__(self, name):
        self.name = name
        self.__books = []

    def add_book(self, book):
        self.__books.append(book)
        print(f"✅ '{book}' added!")

    def show_books(self):
        print(f"\n📚 {self.name} Books:")
        for b in self.__books:
            print(f"  - {b}")

    def total(self):
        return len(self.__books)

lib = Library("City Library")
lib.add_book("Python Basics")
lib.add_book("OOP in Python")
lib.add_book("Data Science")
lib.show_books()
print(f"\nTotal books: {lib.total()}")
```
