## Problem 01 :Student Grade Calculator

```python
def grade():

    s1 = int(input("Enter Subject 1 Marks: "))
    s2 = int(input("Enter Subject 2 Marks: "))
    s3 = int(input("Enter Subject 3 Marks: "))
    s4 = int(input("Enter Subject 4 Marks: "))
    s5 = int(input("Enter Subject 5 Marks: "))

    total = s1 + s2 + s3 + s4 + s5
    average = total / 5
    percentage = total / 500 * 100

    print("Total =", total)
    print("Average =", average)
    print("Percentage =", percentage,"%")

    if percentage >= 90:
        print("Grade = A")
    elif percentage >= 80:
        print("Grade = B")
    elif percentage >= 70:
        print("Grade = C")
    elif percentage >= 60:
        print("Grade = D")
    elif percentage >= 50:
        print("Grade = E")
    else:
        print("Grade = F")
```

```python
grade()
```

## Problem 02 :Number Analyzer

```python
def number_analyzer():

    n = int(input("Enter a Number: "))

    if n > 0:
        print("Positive")
    elif n < 0:
        print("Negative")
    else:
        print("Zero")

    if n % 2 == 0:
        print("Even")
    else:
        print("Odd")
```

```python
number_analyzer()
```

## Problem 03: Simple Calculator

```python
def calculator():
    num1 = int(input("Enter first number: "))
    num2 = int(input("Enter second number: "))

    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")

    choice = input("Enter your choice (1-4): ")

    if choice == "1":
        print("Answer =", num1 + num2)

    elif choice == "2":
        print("Answer =", num1 - num2)

    elif choice == "3":
        print("Answer =", num1 * num2)

    elif choice == "4":
        if num2 != 0:
            print("Answer =", num1 / num2)
        else:
            print("Cannot divide by zero")

    else:
        print("Invalid Choice")
```

```python
calculator()
```

## Problem 04: List Statistics

```python
def list_statistics():
    numbers = []

    for i in range(5):
        n = int(input("Enter Number: "))
        numbers.append(n)

    print("List =", numbers)
    print("Largest =", max(numbers))
    print("Smallest =", min(numbers))
    print("Sum =", sum(numbers))
    print("Average =", sum(numbers) / len(numbers))

    numbers.sort()
    print("Second Largest =", numbers[-2])
```

```python
list_statistics()
```

## Problem 05: Student Record Management System

```python
def student_record():
    students = {}

    while True:
        print("\n1. Add Student")
        print("2. Update Marks")
        print("3. Delete Student")
        print("4. Search Student")
        print("5. Display All Records")
        print("6. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            name = input("Enter Student Name: ")
            marks = int(input("Enter Marks: "))
            students[name] = marks

        elif choice == "2":
            name = input("Enter Student Name: ")
            if name in students:
                marks = int(input("Enter New Marks: "))
                students[name] = marks
            else:
                print("Student Not Found")

        elif choice == "3":
            name = input("Enter Student Name: ")
            if name in students:
                del students[name]
                print("Student Deleted")
            else:
                print("Student Not Found")

        elif choice == "4":
            name = input("Enter Student Name: ")
            if name in students:
                print("Marks =", students[name])
            else:
                print("Student Not Found")

        elif choice == "5":
            print(students)

        elif choice == "6":
            print("Program Ended")
            break

        else:
            print("Invalid Choice")
```

```python
student_record()
```

## Problem 06 : Word Frequency Counter 

```python
def word_count():

    sentence = input("Enter a Sentence: ")

    words = sentence.split()

    count = {}

    for word in words:
        if word in count:
            count[word] = count[word] + 1
        else:
            count[word] = 1

    print(count)
```

```python
word_count()
```

## Problem 08: Shopping Cart

```python
def shopping_cart():

    prices = {
        "apple": 100,
        "banana": 50,
        "milk": 200,
        "bread": 80
    }

    cart = {}

    while True:
        print("\n1. Add Item")
        print("2. Remove Item")
        print("3. Update Quantity")
        print("4. Display Cart")
        print("5. Calculate Bill")
        print("6. Exit")

        choice = input("Enter Choice: ")

        if choice == "1":
            item = input("Enter Item: ")
            qty = int(input("Enter Quantity: "))
            cart[item] = qty

        elif choice == "2":
            item = input("Enter Item to Remove: ")
            if item in cart:
                del cart[item]
                print("Item Removed")
            else:
                print("Item Not Found")

        elif choice == "3":
            item = input("Enter Item: ")
            if item in cart:
                qty = int(input("Enter New Quantity: "))
                cart[item] = qty
            else:
                print("Item Not Found")

        elif choice == "4":
            print("Shopping Cart =", cart)

        elif choice == "5":
            total = 0
            for item in cart:
                total = total + prices[item] * cart[item]
            print("Total Bill =", total)

        elif choice == "6":
            print("Thank You!")
            break

        else:
            print("Invalid Choice")
```

```python
shopping_cart()
```

## Problem 09: Exception Handling

```python
def divide():

    try:
        a = int(input("Enter First Number: "))
        b = int(input("Enter Second Number: "))

        print("Answer =", a / b)

    except ZeroDivisionError:
        print("Cannot Divide by Zero")

    except ValueError:
        print("Invalid Input")
```

```python
divide()
```

## Problem 10: Library Management

```python
def library():

    books = []

    while True:

        print("\n===== Library Menu =====")
        print("1. Add Book")
        print("2. Search Book")
        print("3. Borrow Book")
        print("4. Return Book")
        print("5. Display Books")
        print("6. Exit")

        try:
            choice = int(input("Enter Choice: "))

            if choice == 1:
                book = input("Enter Book Name: ")
                books.append(book)
                print("Book Added Successfully.")

            elif choice == 2:
                book = input("Enter Book Name: ")

                if book in books:
                    print("Book Found.")
                else:
                    print("Book Not Found.")

            elif choice == 3:
                book = input("Enter Book Name: ")

                if book in books:
                    books.remove(book)
                    print("Book Borrowed Successfully.")
                else:
                    print("Book Not Available.")

            elif choice == 4:
                book = input("Enter Book Name: ")
                books.append(book)
                print("Book Returned Successfully.")

            elif choice == 5:
                if len(books) == 0:
                    print("Library is Empty.")
                else:
                    print("Available Books:")
                    for book in books:
                        print(book)

            elif choice == 6:
                print("Thank You!")
                break

            else:
                print("Invalid Choice")

        except:
            print("Enter Numbers Only.")
```

```python
library()
```
